```python
df = pd.read_csv('2) Stock Prices Data Set.csv')
df.head()
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df.set_index('date', inplace=True)
df.head()
```

```python
plt.figure(figsize=(10,5))
plt.plot(df['close'])

plt.title("Stock Price Over Time")
plt.xlabel("date")
plt.ylabel("close Price")

plt.show()
```

```python
result = seasonal_decompose(df['close'], model='additive', period=30)

result.plot()
plt.show()
```

```python
df['Moving_Avg'] = df['close'].rolling(window=30).mean()
```

```python
plt.figure(figsize=(10,5))

plt.plot(df['close'], label='Original Price')
plt.plot(df['Moving_Avg'], label='Moving Average')

plt.title("Stock Price with Moving Average")
plt.xlabel("date")
plt.ylabel("Price")

plt.legend()
plt.show()
```
